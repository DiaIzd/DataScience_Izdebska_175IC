```
import numpy as np
import pandas as pd

#Definicja funkcji aktywacyjnej
def activation(x):
  if x>= 0:
    return 1
  return 0
#Bramka AND
for i in range(0,5):
  if i==1:
    x1 = 0
    x2 = 0
  if i==2:
    x1 = 0
    x2 = 1
  if i==3:
    x1 = 1
    x2 = 0
  if i==4:
    x1 = 1
    x2 = 1
  if i != 0:
    waga_x1 = 1
    waga_x2 = 1 
    bias = 1
    waga_bias = -1.5
    suma = x1*waga_x1 + x2*waga_x2 + bias*waga_bias
    act = activation(suma)
    print(''+str(x1)+' ||'+str(x2)+' || '+str(x1 and x2)+" ||"+str(suma)+"||"+str(act))
  if i==0:
    print("Bramka AND")
    print("----------------------------------")
    print('x1||x2||x1 and x2||(-1.5+x1+x2)||act')
    print("----------------------------------")
  
#Bramka NOT
for i in range(0,3):
  if i == 1:
    x1 == 0
  if i == 2:
    x1 == 1
  if i != 0:
    waga_x1 = -2
    bias = 1
    waga_bias = 1.5
    suma = x1*waga_x1 + bias*waga_bias
    act = activation(suma)
    print(''+str(x1)+' ||'+str(x2)+' || '+str(x1 and x2)+" ||"+str(suma)+" || "+str(act))
  if i==0:
    print("Bramka NOT")
    print("----------------------------------")
    print('x1||x2||x1 or x2||(-0.5+x1+x2)||act')
    print("----------------------------------")
#Bramka OR
for i in range(0,5):
  if i==1:
    x1 = 0
    x2 = 0
  if i==2:
    x1 = 0
    x2 = 1
  if i==3:
    x1 = 1
    x2 = 0
  if i==4:
    x1 = 1
    x2 = 1
  if i != 0:
    waga_x1 = 1
    waga_x2 = 1 
    bias = 1
    waga_bias = -0.5
    suma = x1*waga_x1 + x2*waga_x2 + bias*waga_bias
    act = activation(suma)
    print(''+str(x1)+' || '+str(not x1)+" ||"+str(suma)+"||"+str(act))
  if i==0:
    print("Bramka OR")
    print("----------------------------------")
    print('x1||not x1||(1-2x1)||act')
    print("----------------------------------")
#Bramka XOR 
for i in range(0,5):
  if i==1:
    x1 = 0
    x2 = 0
  if i==2:
    x1 = 0
    x2 = 1
  if i==3:
    x1 = 1
    x2 = 0
  if i==4:
    x1 = 1
    x2 = 1
  if i != 0:
    neuron_OR = activation(x1+x2-0.5)
    not_NOT = -2
    neuron_NOT =activation(x1*not_NOT+1)
    neuron1_NOT =activation(x2*not_NOT+1)
    neuron1_OR = activation(neuron_NOT + neuron1_NOT -0.5)
    neuron_AND = activation(neuron_OR +neuron1_OR-1.5)
   
    print(''+str(x1)+' ||'+str(x2)+" ||"+str(neuron_AND))
  if i==0:
    print("Bramka XOR")
    print("----------")
    print('x1||x2||act')
    print("----------")
#Bramka XNOR
for i in range(0,5):
  if i==1:
    x1 = 0
    x2 = 0
  if i==2:
    x1 = 0
    x2 = 1
  if i==3:
    x1 = 1
    x2 = 0
  if i==4:
    x1 = 1
    x2 = 1
  if i != 0:
    neuron_OR = activation(x1+x2-0.5)
    not_NOT = -2
    neuron_NOT =activation(x1*not_NOT+1)
    neuron1_NOT =activation(x2*not_NOT+1)
    neuron1_OR = activation(neuron_NOT + neuron1_NOT -0.5)
    neuron_AND = activation(neuron_OR +neuron1_OR-1.5)
    neuron2_NOT = activation(neuron_AND*not_NOT+1)
    print(''+str(x1)+' ||'+str(x2)+" ||"+str(neuron2_NOT))
  if i==0:
    print("Bramka XNOR")
    print("----------")
    print('x1||x2||act')
    print("----------")
```
